```python
import textgrid
```

```python
tg = textgrid.TextGrid()
```

```python
interval_tier = textgrid.IntervalTier(name="words", minTime=0, maxTime=5)
```

```python
# Add intervals to the interval tier
interval_tier.add(0.0, 1.0, "Hello")
interval_tier.add(1.0, 2.0, "world")
interval_tier.add(2.0, 3.5, "this is")
interval_tier.add(3.5, 5.0, "a test")
```

```python
tg.append(interval_tier)
```

```python
tg.write("sample.TextGrid")
```

```python
loaded_tg = textgrid.TextGrid.fromFile("sample.TextGrid")
```

```python
loaded_tg.
```

```python
words_tier = loaded_tg.getFirst("words")
for interval in words_tier:
    print(f"Start: {interval.minTime}, End: {interval.maxTime}, Text: {interval.mark}")
```

```python
words_tier[1].mark = "hello world"
```

```python
loaded_tg.write("modified_example.TextGrid")
```
